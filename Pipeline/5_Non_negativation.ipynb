## Functions

```python
def shift_column(df: pd.DataFrame, col: str):
    col_min = min(df[col])
    df[col] = df[col].apply(lambda x: x - col_min)
```


## File selection

```python
file = widgets.Dropdown(options=os.listdir(input_root), description='Dataset')

display(file)
```

# Data shifting

## Dataset loading

```python
path = os.path.join(input_root, file.value)

with Halo(text=f'Reading {path}...', spinner='dots'):
    df = pd.read_pickle(path)
df
```

## Negative columns research

```python
with Halo(text='Searching negative columns...', spinner='dots'):
    negative_cols = df.lt(0).any()[lambda x: x].index

if len(negative_cols) == 0:
    print('All columns are positives')
else:
    print(f'{len(negative_cols)} negative columns found')
```

## Shifting

```python
with Halo(text='Shifting values...', spinner='dots'):

    shifted = df.copy()

    shifted[negative_cols] = df[negative_cols] - df[negative_cols].min(axis=0)

shifted
```

## Output to pickle

```python
path = os.path.join(output_root, file.value)

with Halo(text=f'Writing {path}...', spinner='dots'):
    shifted.to_pickle(path)
print('Done')
```
